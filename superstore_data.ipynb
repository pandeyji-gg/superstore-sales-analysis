```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series as se
from pandas import DataFrame as df
```

```python
data=pd.read_csv('train.csv')
```

```python
frame=df(data)
frame.head()
```

```python
region_total_sales=frame.groupby('Region')['Sales'].sum()
ax=region_total_sales.sort_values(ascending=True).plot(kind='barh',figsize=(10,6))
for p in ax.patches:
    ax.annotate(f'{p.get_width():,.0f}', 
                (p.get_width(), p.get_y() + p.get_height()/2),
                va='center')
plt.xlim(0,800000)
plt.title("Total sales per region")
plt.tight_layout()
plt.show()
```

```python
category_total_sales=frame.groupby('Category')['Sales'].sum()
ax=category_total_sales.sort_values(ascending=True).plot(kind='barh',figsize=(10,6))
for p in ax.patches:
    ax.annotate(f'{p.get_width():,.0f}', 
                (p.get_width(), p.get_y() + p.get_height()/2),
                va='center')
plt.xlim(0,900000)
plt.title("Total sales per category")
plt.tight_layout()
plt.show()
```

```python
frame1=frame[frame['Category']=='Technology']
technology_by_subcategory=frame1.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
ax=technology_by_subcategory.plot(kind='barh',figsize=(10,5))
for p in ax.patches:
    ax.annotate(f'{p.get_width():,.0f}', 
                (p.get_width(), p.get_y() + p.get_height()/2),
                va='center')
plt.xlim(0,4000000)
plt.title("Total sales per sub-category of Technology")
plt.tight_layout()
plt.show()
```

```python
frame['Order Date'] = pd.to_datetime(frame['Order Date'],dayfirst=True)
frame['Month'] = frame['Order Date'].dt.month
frame['Year']=frame['Order Date'].dt.year
for_each_month=frame.groupby(['Year','Month'])['Sales'].sum()
for_each_month = for_each_month.reset_index()
for_each_month['Date'] = pd.to_datetime(for_each_month[['Year','Month']].assign(Day=1))
for_each_month = for_each_month.set_index('Date')['Sales']
for_each_month.plot(kind='line',figsize=(15,5))
plt.title("monthly sales trend 2015-2018")
plt.xlabel('Year, Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
